```python
import pandas as pd
import matplotlib.pyplot as plt

batting_df = pd.read_csv("/Users/gamingspectrum24/Documents/University Coursework/6th Semester/Cognitive Analytics/Lab/End-to-End-Cricket-Data-Analytics/Scraped data/t20_csv_files/batting_summary.csv")

bowling_df = pd.read_csv("/Users/gamingspectrum24/Documents/University Coursework/6th Semester/Cognitive Analytics/Lab/End-to-End-Cricket-Data-Analytics/Scraped data/t20_csv_files/bowling_summary.csv")

players_info = pd.read_csv("/Users/gamingspectrum24/Documents/University Coursework/6th Semester/Cognitive Analytics/Lab/End-to-End-Cricket-Data-Analytics/Scraped data/t20_csv_files/players_info_no_images.csv")
```

```python
best_batting = batting_df.sort_values(by="runs", ascending=False).head(5)
best_bowling = bowling_df.sort_values(by="wickets", ascending=False).head(5)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

best_batting.plot.bar(x="batsmanName", y="runs", ax=axes[0], color="skyblue")
axes[0].set_title("Top Batsmen by Runs")
axes[0].set_ylabel("Runs")

best_bowling.plot.bar(x="bowlerName", y="wickets", ax=axes[1], color="lightgreen")
axes[1].set_title("Top Bowlers by Wickets")
axes[1].set_ylabel("Wickets")

plt.tight_layout()
plt.show()
```

```python
print(players_info.dtypes)

# Convert columns from string to int if necessary
if players_info['runs'].dtype == 'object':
    players_info['runs'] = players_info['runs'].astype(int)

# Create a bar chart to visualize the runs scored by players
plt.figure(figsize=(10, 6))
plt.bar(players_info['batsmanName'], players_info['runs'])
plt.xlabel('Batsman Name')
plt.ylabel('Runs')
plt.title('Runs Scored by Players')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(14, 10))
plt.subplots_adjust(hspace=0.5, wspace=0.5)

for idx, player in enumerate(players_info):
    row_idx = idx // 4
    col_idx = idx % 4
    ax = axes[row_idx, col_idx]
    ax.axis('off')
    ax.text(0.5, 0.5, f"Name: {player['name']}\nTeam: {player['team']}\nBatting Style: {player['battingStyle']}\nBowling Style: {player['bowlingStyle']}\nPlaying Role: {player['playingRole']}",
            fontsize=10, ha='center', va='center', wrap=True)

plt.show()
```
